1.What is unsupervised learning in the context of machine learning?

.Unsupervised learning is a type of machine learning where the model learns patterns from unlabeled data—meaning there are no predefined outputs. It’s mainly used for clustering, dimensionality reduction, and anomaly detection, helping discover hidden structures or relationships in the data without supervision.

2.How does K-Means clustering algorithm work?

. K-Means clustering works by partitioning data into K clusters based on feature similarity:

Initialize: Randomly choose K centroids (cluster centers).

Assign: Each data point is assigned to the nearest centroid (based on distance).

Update: Recalculate centroids as the mean of all points in each cluster.

Repeat: Steps 2 and 3 until centroids stabilize (no change or minimal change

3.Explain the concept of a dendrogram in hierarchical clustering.

.A dendrogram is a tree-like diagram used in hierarchical clustering to show how data points are merged or split at different levels of similarity. Each branch represents a cluster, and the height at which two branches join indicates the distance or dissimilarity between those clusters. It helps visualize the clustering process and decide the optimal number of clusters by cutting the tree at a specific height.

4.What is the main difference between K-Means and Hierarchical Clustering?

. The main difference is:

K-Means requires specifying the number of clusters (K) in advance and partitions data into non-overlapping, flat clusters.

Hierarchical Clustering builds a tree-like structure (dendrogram) and does not require a predefined number of clusters, allowing clusters to be nested and merged at different levels.

5.What are the advantages of DBSCAN over K-Means

. Advantages of DBSCAN over K-Means:

Can find clusters of any shape (not just spherical like K-Means).

Doesn’t require specifying the number of clusters beforehand.

Can identify and label noise/outliers as points not belonging to any cluster.

Works well with clusters of varying densities.

6.When would you use Silhouette Score in clustering?

. You use the Silhouette Score to evaluate how well clusters are formed—it measures how similar an object is to its own cluster compared to other clusters. It helps:

Assess cluster quality (higher scores mean better-defined clusters).

Choose the optimal number of clusters by comparing scores for different K values.

Compare different clustering algorithms on the same dataset.

7.What are the limitations of Hierarchical Clustering?

. Limitations of Hierarchical Clustering:

Computationally expensive and slow for large datasets.

Does not scale well to high-dimensional data.

Once merged or split, clusters cannot be undone (no reassignments).

Sensitive to noise and outliers which can distort the hierarchy.

Requires a distance metric that might not suit all data types.

8.Why is feature scaling important in clustering algorithms like K-Means?

. Feature scaling is important in K-Means clustering because:

K-Means uses distance measures (usually Euclidean distance) to assign points to clusters.

If features have different scales, those with larger ranges will dominate the distance calculation, biasing the clustering.

Scaling ensures all features contribute equally, leading to more meaningful clusters.

9.How does DBSCAN identify noise points?

. DBSCAN identifies noise points as data points that do not belong to any cluster because they have fewer than the minimum required neighbors (minPts) within the specified radius (eps). These points are considered outliers or noise.

10.Define inertia in the context of K-Means.

. Inertia in K-Means is the sum of squared distances between each data point and its assigned cluster centroid. It measures how tightly the clusters are formed — lower inertia means more compact clusters.

11.What is the elbow method in K-Means clustering?

. he elbow method helps choose the optimal number of clusters (K) in K-Means by:

Plotting inertia (sum of squared distances) vs. different K values.

Looking for the "elbow" point where inertia starts decreasing more slowly.

This point balances cluster compactness and simplicity, indicating a good K.

12.Describe the concept of "density" in DBSCAN

.In DBSCAN, density refers to the number of data points within a specified radius (eps) around a point. A point is considered a core point if it has at least a minimum number (minPts) of neighbors within this radius. Clusters form by connecting core points and their neighbors, reflecting areas of high point density.

13.Can hierarchical clustering be used on categorical data.

. Yes, hierarchical clustering can be used on categorical data, but it requires an appropriate distance or similarity measure for categorical variables (like Hamming distance or Jaccard similarity). Unlike numeric data, you can’t use Euclidean distance directly, so choosing the right metric is key.

14.What does a negative Silhouette Score indicate?

. A negative Silhouette Score indicates that a data point is misclassified, meaning it is closer to points in a different cluster than to points in its own cluster. Overall, a negative average score suggests poor clustering quality and overlapping or incorrect clusters.

15.Explain the term "linkage criteria" in hierarchical clustering?

. Linkage criteria in hierarchical clustering define how the distance between clusters is calculated when merging them. Common types include:

Single linkage: distance between the closest points of two clusters.

Complete linkage: distance between the farthest points.

Average linkage: average distance between all pairs of points.

Ward’s linkage: minimizes the variance within clusters after merging.

16.Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

. K-Means clustering can perform poorly on data with varying cluster sizes or densities because:

It assumes clusters are spherical and equally sized, which may not match real data.

It uses centroids and Euclidean distance, so:

Larger or denser clusters can dominate.

17.What are the core parameters in DBSCAN, and how do they influence clustering?

. The core parameters in DBSCAN are:

eps: Maximum distance to consider points as neighbors.

minPts: Minimum number of neighbors to form a dense region.

They influence how dense a region must be to form a cluster—smaller eps or larger minPts lead to fewer, tighter clusters and more noise.

18.How does K-Means++ improve upon standard K-Means initialization?

. K-Means++ improves standard K-Means by smarter initialization of centroids. Instead of random selection, it:

Chooses the first centroid randomly.

Selects the next centroids with a probability proportional to the square of the distance from the nearest existing centroid.

19.What is agglomerative clustering?

. Agglomerative clustering is a bottom-up hierarchical clustering method where each data point starts as its own cluster, and pairs of clusters are merged step by step based on similarity until all points form one cluster or a stopping condition is met.

20.What makes Silhouette Score a better metric than just inertia for model evaluation?

. Silhouette Score is better than inertia because:

It considers both intra-cluster cohesion (how close points are in the same cluster) and inter-cluster separation (how far they are from other clusters).

It gives a normalized score (-1 to 1), making it easier to compare models.

Inertia only measures compactness, not how well-separated clusters are.

21.Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a

29.Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart


```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)

sil_scores = []
k_values = range(2, 6)

# Calculate silhouette scores for k = 2 to 5
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    sil_scores.append(score)

# Plot silhouette scores
plt.figure(figsize=(8, 5))
plt.bar(k_values, sil_scores, color='skyblue')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for K-Means Clustering')
plt.xticks(k_values)
plt.grid(axis='y')
plt.show()
```

31.Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with overlapping clusters by increasing cluster_std
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=2.5, random_state=42)

# Fit K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot decision boundaries
h = 0.1  # step size in the mesh

# Create mesh grid
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict cluster for each point in mesh
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plotting
plt.figure(figsize=(10, 7))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, marker='X', label='Centroids')
plt.title('K-Means Clustering with Overlapping Clusters and Decision Boundaries')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

34.Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate noisy circular data
X, y_true = make_circles(n_samples=300, factor=0.5, noise=0.05, random_state=42)

# Scale data (important for DBSCAN)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
labels_kmeans = kmeans.fit_predict(X_scaled)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels_dbscan = dbscan.fit_predict(X_scaled)

# Plot results side-by-side
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# KMeans plot
axs[0].scatter(X[:, 0], X[:, 1], c=labels_kmeans, cmap='viridis', s=50, edgecolor='k')
axs[0].set_title('KMeans Clustering')
axs[0].set_xlabel('Feature 1')
axs[0].set_ylabel('Feature 2')
axs[0].grid(True)

# DBSCAN plot
# Mark noise points (label = -1) in red
colors = ['red' if label == -1 else 'blue' for label in labels_dbscan]
axs[1].scatter(X[:, 0], X[:, 1], c=colors, s=50, edgecolor='k')
axs[1].set_title('DBSCAN Clustering')
axs[1].set_xlabel('Feature 1')
axs[1].set_ylabel('Feature 2')
axs[1].grid(True)

plt.suptitle('Comparison of KMeans and DBSCAN on Noisy Circular Data')
plt.show()

```

44.Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering


```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

# Load Iris dataset
iris = load_iris()
X = iris.data

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Calculate Silhouette Coefficients for each sample
silhouette_vals = silhouette_samples(X, labels)

# Sort the samples by silhouette score within each cluster
y_lower = 10
plt.figure(figsize=(8, 6))
for i in range(3):
    ith_cluster_silhouette_vals = silhouette_vals[labels == i]
    ith_cluster_silhouette_vals.sort()
    size_cluster_i = ith_cluster_silhouette_vals.shape[0]
    y_upper = y_lower + size_cluster_i

    plt.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_vals,
                      alpha=0.7)
    plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10  # 10 for spacing between clusters

plt.xlabel("Silhouette Coefficient Values")
plt.ylabel("Cluster Label")
plt.title("Silhouette Plot for KMeans Clustering on Iris Dataset")
plt.axvline(x=silhouette_score(X, labels), color="red", linestyle="--", label="Average Silhouette Score")
plt.legend()
plt.show()
```

46.Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4
features).

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd

# Load Wine dataset
wine = load_wine()
X = wine.data
feature_names = wine.feature_names

# Use only the first 4 features
X_subset = X[:, :4]

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X_subset)

# Create a DataFrame for seaborn
df = pd.DataFrame(X_subset, columns=feature_names[:4])
df['Cluster'] = labels.astype(str)  # for color coding in pairplot

# Plot pairplot with cluster hues
sns.pairplot(df, hue='Cluster', palette='Set1', diag_kind='kde', plot_kws={'alpha':0.7})
plt.suptitle('KMeans Cluster Assignments on Wine Dataset (First 4 Features)', y=1.02)
plt.show()
```

47.Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np

# Generate noisy blobs
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=2.5, random_state=42)

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Count clusters (excluding noise label -1)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f'Number of clusters found: {n_clusters}')
print(f'Number of noise points: {n_noise}')
```
